```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import preprocessing
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

### Импортируем данные двух операторов
#### Дала-Телеком

```python
dala_telecom_df = pd.read_csv('dala_telecom.csv', delimiter=';')
print("Кол-во записей в датасете:", dala_telecom_df.shape)
print(dala_telecom_df.info())
print(dala_telecom_df.isnull().sum())
dala_telecom_df["SESSION_DURATION_IN_SEC"] = dala_telecom_df["SESSION_DURATION_IN_SEC"].fillna(0)
dala_telecom_df["PAY_AMT"] = dala_telecom_df["PAY_AMT"].fillna(0)
dala_telecom_users_count = dala_telecom_df.nunique()['ID']
print(f"Кол-во пользователей в базе Дала-Телеком: {dala_telecom_users_count}")
dala_telecom_df.head()
```

#### Кен-Телеком

```python
ken_telecom_df = pd.read_csv('ken_telecom.csv', delimiter=';')
print("Кол-во записей в датасете:", ken_telecom_df.shape)
print(ken_telecom_df.info())
print(ken_telecom_df.isnull().sum())
ken_telecom_df["SESSION_DURATION_IN_SEC"] = ken_telecom_df["SESSION_DURATION_IN_SEC"].fillna(0)
ken_telecom_df["PAY_AMT"] = ken_telecom_df["PAY_AMT"].fillna(0)
ken_telecom_users_count = ken_telecom_df.nunique()['ID']
print(f"Кол-во пользователей в базе Кен-Телеком: {ken_telecom_users_count}")
ken_telecom_df.head()
```

### Ответ 1 a)
#### Сначала выбираем сессии с типом voice и продолжительностью > 0 у двух операторов, считаем среднее значение продолжительности сессии и находим абсолютную разницу между найденными значениями

```python
dala_voice = dala_telecom_df[(dala_telecom_df.TYPE_OF_CONNECTION == 'voice') & (dala_telecom_df.SESSION_DURATION_IN_SEC > 0)]
dala_mean_duration = dala_voice["SESSION_DURATION_IN_SEC"].mean()

ken_voice = ken_telecom_df[(ken_telecom_df.TYPE_OF_CONNECTION == 'voice') & (ken_telecom_df.SESSION_DURATION_IN_SEC > 0)]
ken_mean_duration = ken_voice["SESSION_DURATION_IN_SEC"].mean()
diff = round(abs(dala_mean_duration - ken_mean_duration), 1)
print(f"Абсолютная разница между средней продолжительностью состоявшихся голосовых исходящих сессий \ 
        абонентских баз мобильных операторов конкурентов: {diff}")
```

### Ответ 1 b)
#### Сначала выбираем сессии с типом gprs у оператора Кен-Телеком

```python
ken_gprs = ken_telecom_df[ken_telecom_df.TYPE_OF_CONNECTION == 'gprs'].copy()
ken_gprs.reset_index(drop=True)
ken_gprs.head()
```

#### Чтобы выбрать самый нагруженный час, из даты начала сессии выбираем дату и час, группируем по дате и часу и считаем кол-во сессий соответственно. Далее выбираем час и дату , в которые было совершено макс. кол-во сессий. Затем считаем кол-во абонентов, совершавших сессии в найденный час и дату (кол-во уникальных ID) и находим отношение к общему кол-во абонентов.
##### P.s. считаю нагрузкой кол-во совершенных сессий в данный час.

```python
ken_gprs['SESSION_START_TIME'] = pd.to_datetime(ken_gprs['SESSION_START_TIME'])
ken_gprs['SESSION_START_HOUR'] = ken_gprs['SESSION_START_TIME'].dt.hour
ken_gprs['SESSION_START_DATE'] = ken_gprs['SESSION_START_TIME'].dt.date

ken_gprs_max = ken_gprs.groupby(['SESSION_START_DATE', 'SESSION_START_HOUR']).size()
    .sort_values(ascending=False)
    .reset_index(name='count').iloc[0]
print(f"Самый нагруженный час {ken_gprs_max['SESSION_START_DATE']} в {ken_gprs_max['SESSION_START_HOUR']} \
        с кол-вом сессий {ken_gprs_max['count']}")

ken_gprs_busiest = ken_gprs[(ken_gprs.SESSION_START_DATE == ken_gprs_max['SESSION_START_DATE']) & 
                            (ken_gprs.SESSION_START_HOUR == ken_gprs_max['SESSION_START_HOUR'])]

ken_gprs_busiest_users_count = ken_gprs_busiest.nunique()['ID']
percent = round(ken_gprs_busiest_users_count / ken_telecom_users_count * 100, 1)

print(f"Кол-во абонентов Кен_телеком {ken_telecom_users_count}. \
        Кол-во абонентов в самый нагруженный час {ken_gprs_busiest_users_count}. В процентах {percent} %")
```

Далее выбираем час и дату , в которые было совершено мин. кол-во сессий. Затем считаем кол-во абонентов, совершавших сессии в найденный час и дату (кол-во уникальных ID) и находим отношение к общему кол-во абонентов.

```python
dala_voice['SESSION_START_TIME'] = pd.to_datetime(dala_voice['SESSION_START_TIME'])
dala_voice['SESSION_START_HOUR'] = dala_voice['SESSION_START_TIME'].dt.hour
dala_voice['SESSION_START_DATE'] = dala_voice['SESSION_START_TIME'].dt.date

dala_voice_min = ken_gprs.groupby(['SESSION_START_DATE', 'SESSION_START_HOUR']).size()
    .sort_values(ascending=True)
    .reset_index(name='count').iloc[0]
print(f"Самый низконагруженный час {dala_voice_min['SESSION_START_DATE']} в {dala_voice_min['SESSION_START_HOUR']} \
    с кол-вом сессий {dala_voice_min['count']}")

dala_voice_nonbusiest = dala_voice[(dala_voice.SESSION_START_DATE == dala_voice_min['SESSION_START_DATE']) 
                                   & (dala_voice.SESSION_START_HOUR == dala_voice_min['SESSION_START_HOUR'])]

dala_voice_nonbusiest_users_count = dala_voice_nonbusiest.nunique()['ID']
percent = round(dala_voice_nonbusiest_users_count / dala_telecom_users_count * 100, 1)

print(f"Кол-во абонентов Дала_телеком {dala_telecom_users_count}. \
    Кол-во абонентов в самый ненагруженный час {dala_voice_nonbusiest_users_count}. В процентах {percent} %")
```

#### Импортируем файл счетчиков по абонентам

```python
abon_profile_df = pd.read_csv('abon_profile.csv', delimiter=';')
abon_profile_df.head()
```

```python
def getClusteredAbonDf(df, clusterNum):
    X = df.values[:,1:]
    X = np.nan_to_num(X)
    Clus_dataSet = StandardScaler().fit_transform(X)
    #clusterNum = 2
    k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
    k_means.fit(Clus_dataSet)
    labels = k_means.labels_
    df_clust = df.copy()
    df_clust["Cluster"] = labels
    return df_clust
```

#### В предыдущей функции я заменял пустые ячейки на 0, теперь попробовал заменить на среднее значение. Результаты отличаются, но не намного.

```python
def getClusteredAbonDf2(df, clusterNum):
    X = df.values[:,1:]
    #X = np.nan_to_num(X)
    missing = ~np.isfinite(X)
    mu = np.nanmean(X, 0, keepdims=1)
    X = np.where(missing, mu, X)
    Clus_dataSet = StandardScaler().fit_transform(X)
    #clusterNum = 2
    k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
    k_means.fit(Clus_dataSet)
    labels = k_means.labels_
    df_clust = df.copy()
    df_clust["Cluster"] = labels
    return df_clust
```

```python
ken_telecom_abon_profile_clust2 = getClusteredAbonDf2(ken_abon_profile_df, 2)
ken_telecom_abon_profile_clust2.groupby('Cluster').mean()
```

```python
ken_telecom_abon_profile_clust2['Cluster'].value_counts()
```

```python
def create_flat_table(df):
    df_flat = df.groupby('ID').agg({'PAY_AMT' : 'sum', 'SESSION_DURATION_IN_SEC' : 'sum', 'SERVICE_ID' : lambda x: x.value_counts().index[0]})
    df_flat = df_flat.rename(columns={'PAY_AMT':'TOTAL_PAY_AMT', 'SESSION_DURATION_IN_SEC':'TOTAL_SESSION_DURATION_IN_SEC', 'SERVICE_ID':'FAVOURITE_SERVICE_ID'})
    df_flat = pd.merge(df_flat, abon_profile_df, on='ID', how='left')
    return df_flat
```
